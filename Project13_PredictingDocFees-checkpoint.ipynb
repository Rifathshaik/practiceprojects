# Outlier Removal:

```python
from scipy.stats import zscore

z=np.abs(zscore(train_data))

print(np.where(z>3))
```

```python
df1=train_data[(z<3).all(axis=1)]
print("with outliers::",train_data.shape)
print("After removing outliers::",df1.shape)
```

# Skewness Check:

```python
df1.skew()
```

```python
#splitting data into X and Y for model building

X=df1.drop(['Fees'],axis=1)
Y=df1['Fees']
```

```python
#CHECKING SHAPE OF X AND Y

X.shape
```

```python
Y.shape
```

```python
Y=Y.values.reshape(-1,1)
```

```python
Y.shape
```

```python
#SCALING X VALUES

from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
X=sc.fit_transform(X)
```

```python
#skewness removal

from sklearn.preprocessing import power_transform
X=power_transform(X,method='yeo-johnson')
```

```python
#Model building

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.linear_model import SGDRegressor
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
import warnings
warnings.simplefilter('ignore')
```

```python
best_Randstate=0
accuracy=0
model=[LinearRegression(),RandomForestRegressor(),AdaBoostRegressor(),SGDRegressor(),Lasso(),Ridge()]
for m in model:
            for i in range(20,100):
                                    x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=.25,random_state=i)
                                    m.fit(x_train,y_train)
                                    predicted_y=m.predict(x_test)
                                    accu_temp=r2_score(y_test,predicted_y)
                                    if(accu_temp>accuracy):
                                                accuracy=accu_temp
                                                best_Randstate=i
            print("Acuuracy of ",accuracy,"for Model::",m,"found in Random state:",best_Randstate)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=90)
```

```python
rfr=RandomForestRegressor()

rfr_score=cross_val_score(rfr,X,Y,cv=5,scoring='r2')

print("cross validation score:",np.mean(rfr_score))


```

```python
rfr.fit(x_train,y_train)
predicted_rfr=rfr.predict(x_test)
print("Accuracy score::",r2_score(y_test,predicted_rfr))
print("Training accuracy::",rfr.score(x_train,y_train))
print("Test accuracy::",rfr.score(x_test,y_test))
```

# ENSEMBLE TECHNIQUES:

```python
adb=AdaBoostRegressor()

adb_score=cross_val_score(adb,X,Y,cv=10,scoring='r2')

print("cross validation score:",np.mean(adb_score))


```

```python
adb.fit(x_train,y_train)
predicted_adb=adb.predict(x_test)
print("Accuracy score::",r2_score(y_test,predicted_adb))
print("Training accuracy::",adb.score(x_train,y_train))
print("Test accuracy::",adb.score(x_test,y_test))
```

# HYPER PARAMETER TUNING:

```python
parameters={'n_estimators': range(10,50),
            'loss':('linear', 'square', 'exponential'),
            'random_state': range(10,100)
    
}
```

```python
from  sklearn.model_selection import GridSearchCV

adb_grid=GridSearchCV(AdaBoostRegressor(),parameters,cv=5,scoring='r2')
```

```python
adb_grid.fit(x_train,y_train)
adb_pred=adb_grid.best_estimator_.predict(x_test)
print("Accuracy after parameter tuning::",r2_score(y_test,adb_pred))
```
