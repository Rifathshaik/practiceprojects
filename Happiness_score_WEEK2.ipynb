# Removing Outliers

```python
from scipy import stats
import numpy as np
z = np.abs(stats.zscore(df))
print(z)

#defining threshold,values above threshold are outliers
threshold = 3
print(np.where(z > 3))


#removing outliers
df1=df[(z<3).all(axis=1)]
print("with outliers::",df.shape)
print("After removing outliers::",df1.shape)
```

NOTE: only 5% of data is removed by Zscore which can be considered as a good one

# DATA MODELLING

```python
from sklearn.linear_model import LinearRegression
```

```python
x=df1.iloc[:,0:11]
```

```python
y=df1['Dystopia Residual']
```

```python
x.shape
```

```python
x
```

```python
y.shape
```

```python
y
```

Before splitting data into training and testing,reducing the skewness 

```python
from sklearn.preprocessing import power_transform
```

```python
x=power_transform(x,method='yeo-johnson')
```

```python
x.shape
```

```python
y=y.values.reshape(-1,1)
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
best_Randstate=0
accuracy=0
for i in range(20,200):
            x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=i)
            m=LinearRegression()
            m.fit(x_train,y_train)
            predicted_y=m.predict(x_test)
            accu_temp=r2_score(y_test,predicted_y)
            if accu_temp>accuracy :
                    accuracy=accu_temp
                    best_Randstate=i
print("Accuracy of ",accuracy*100,"found in Random state:",best_Randstate)
```

For test size 25, Accuracy of  98.99601763987003 found in Random state->68

proceeding with model building

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=68)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
predict_y=lr.predict(x_test)
accuracy=r2_score(y_test,predict_y)
print("Accuracy of Linear Regression model::",accuracy)
```

```python
print("Accuracy of train data::",lr.score(x_train,y_train))
```

```python
print("Accuracy of test data",lr.score(x_test,y_test))
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print("Mean Absolute Error::",mean_absolute_error(y_test,predict_y))
print("Mean Squared Error::",mean_squared_error(y_test,predict_y))
```

```python
print("Co efficients::\n",lr.coef_)
```

```python
print("Slope::",lr.intercept_)
```
